```python
#take two numbers as input from the user and print their sum, difference, product, and quotient
# each on a separate line. Handle division by zero with an appropriate message.


num1 = int(input("Enter num1: "))
num2 = int(input("Enter num2: "))

sum = num1 + num2
print("Sum =", sum)

diff = num1 - num2
print("Difference =", diff)

product = num1 * num2
print("Product =", product)

if num2 != 0:
    div = num1 / num2
    print("Quotient =", div)
else:
    print("Division by zero is not possible")
```

```python
# Accept a temperature in Celsius from the user, convert it to Fahrenheit and Kelvin,
 #and print all three values with labels.

c = float(input("Enter temperature in Celsius: "))

f = (c * 9/5) + 32
k = c + 273.15

print("Celsius =", c)
print("Fahrenheit =", f)
print("Kelvin =", k)
```

```python
#take a string as input and print: its uppercase version, its length, and the string reversed

s = input("Enter a string: ")

print("Uppercase =", s.upper())
print("Length =", len(s))
print("Reversed =", s[::-1])
```

```python
# Take a year as input and print whether it is a leap year or not.
year = int(input("Enter the year: "))

if year % 4 == 0:

    if year % 100 == 0:

        if year % 400 == 0:
            print("Leap century year")

        else:
            print("Century year but not leap year")

    else:
        print("Leap year but not century year")

else:
    print("Not a leap year")
```

```python
#Build a simple calculator — take two numbers and an operator (+, -, *, /) as input, perform the operation
# and print the result. Print an error message for invalid operators or division by zero.

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

if op == "+":
    print("Result =", num1 + num2)

elif op == "-":
    print("Result =", num1 - num2)

elif op == "*":
    print("Result =", num1 * num2)

elif op == "/":
    if num2 != 0:
        print("Result =", num1 / num2)
    else:
        print("Division by zero is not possible")

else:
    print("Invalid Operator")
```

```python
# Take a number as input and print its multiplication table from 1 to 10. Format each line as: 7 x 3 = 21

n = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{n} x {i} = {n*i}")
```

```python
# Take a positive integer n as input and print all prime numbers from 2 to n.

n = int(input("Enter a number: "))

for num in range(2, n + 1):

    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num)
```

```python
# Keep accepting numbers from the user until they enter 0.
#Then print the total count of numbers entered, their sum, and their average. Do not include 0 in the calculations.

count = 0
sum = 0

num = int(input("Enter a number: "))

while num != 0:
    count += 1
    sum += num
    num = int(input("Enter a number: "))

print("Count =", count)
print("Sum =", sum)

if count > 0:
    print("Average =", sum / count)
```

```python
# Take n as input and print a triangle pattern of n rows made of asterisks, where row i has (2i - 1) stars centered appropriately.

n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```
