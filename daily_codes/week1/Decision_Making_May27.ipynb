```python
#check the number between 50 and 100 or not (use nested if )
```

```python
num = int(input("Enter a number: "))

if num > 50:
    if num < 100:
        print("The number is between 50 and 100")
    else:
        print("The number is greater than 100")
else:
    print("The number is less than or equal to 50")
```

```python
# read puchase cost and selling cost from the user
#state the profit or loss and how much
```

```python
purchase = int(input("Enter the purchase cost: "))
selling = int(input("Enter the selling cost: "))

if selling > purchase:
    print("Profit is", selling - purchase)
else:
    print("Loss is", purchase - selling)
```

```python
#check if the year is leap year or not
year = int(input("Enter the year: "))

if year % 4 == 0:

    if year % 100 == 0:

        if year % 400 == 0:
            print("Leap century year")

        else:
            print("Century year but not leap year")

    else:
        print("Leap year but not century year")

else:
    print("Not a leap year")
```

```python
print("Welcome to the simple calculator")

num = int(input("Enter the number: "))

print("Menu -> 1.Square  2.Cube  3.Check +ve or -ve")

ch = int(input("Select your choice: "))

if ch == 1:
    print(num * num)

elif ch == 2:
    print(num ** 3)

elif ch == 3:

    if num > 0:
        print("num is +ve")

    elif num == 0:
        print("zero")

    else:
        print("num is -ve")

else:
    print("Invalid choice")
```

```python
#menu driven -> take a string from the user
#menu-> 1 upper, 2 lowercase, 3 replace a particular part of the string
#for third option it should which part to replace with what
```

```python
s=input("enter the string=")
print("Menu-> \n 1. upper, \n 2.lower, \n 3. replace")
ch=int(input("enter the choice="))
if ch==1:
  print(s.upper())
elif ch==2:
  print(s.lower())
elif ch==3:
  s1=input("enter the word to replace=")
  s2=input("enter the new word=")
  print(s.replace(s1,s2))
else:
  print("invalid choice")

```
