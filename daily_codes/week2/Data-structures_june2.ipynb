```python
s = input("Enter a string: ")

count = 0
d = {}

for ch in s.lower():
    if ch in "aeiou":
        count += 1

        if ch in d:
            d[ch] += 1
        else:
            d[ch] = 1

print("Total vowels =", count)
print("Vowel counts =", d)
```

```python
s = input("Enter a string: ")

rev = ""

for ch in s:
    rev = ch + rev

if s == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```
